# **2 - Data cleaning**

```python
# Identify and remove duplicates.
print("there is no duplicates")
duplicates = df.duplicated()
df[duplicates]
```

```python
# Identify and deal with missing values in the data.
print(f'we can see that there is no null values in our dataframe. ')
df.isnull().sum()
```

```python
# see how many unique values in our columns

print(f'we can see that Continent column have 6 and thats makes sense (becouse we have only 6 Continents). ')
df.nunique()
```

```python
# order our data by 2022 population in descending order and set Rank as index
df = df.sort_values(by = '2022 Population', ascending = False)
df.set_index('Rank', inplace = True)
df
```

# **3 - Exploratoy Data Analysis**

```python
# make a plot for Population Distribution by Continent
grouped_df = df.groupby('Continent').sum(numeric_only=True).sort_values(by="2022 Population", ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(grouped_df.index, grouped_df['2022 Population'], color='teal', edgecolor='black')
plt.xlabel('Continent')
plt.ylabel('Population (2022)')
plt.title('Bar Population Distribution by Continent 2022')
plt.xticks(rotation=45)
plt.grid(True)

# Adding data labels above each bar
for i, v in enumerate(grouped_df['2022 Population']):
    plt.text(i, v + 500000, str(v), ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
# Plotting a horizontal bar chart for the top 10 countries with the largest population in 2022
top_10_population = df.nlargest(10, '2022 Population')

plt.figure(figsize=(10, 6))
plt.barh(top_10_population['Country/Territory'], top_10_population['2022 Population'], color='steelblue', edgecolor='black')

plt.xlabel('Population (2022)')
plt.title('Top 10 Countries with Largest Population in 2022')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adding data labels
for index, value in enumerate(top_10_population['2022 Population']):
    plt.text(value, index, f'{value:,}', va='center', fontsize=9, color='black', ha='left')

plt.show()
```
