# Preprocessing

## Drop nan values

```python
df.dropna(inplace=True)
```

```python
df.shape
```

# Preprocess Function

```python
# load english language model and create nlp object from it
nlp = spacy.load('en_core_web_sm')
```

```python
# use this utility function to get the preprocessed text data
def preprocess(text):
    # remove stop words and lemmatize the text
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)

    return " ".join(filtered_tokens)
```

# Apply preprocess function on dataframe

```python
df['Preprocessed_Text'] = df['Text'].apply(preprocess)
```

```python
df
```

## Encoding target column


```python
encoder = LabelEncoder()
df['Label_num'] = encoder.fit_transform(df['Label'])
```

```python
df.head(5)
```

## Split data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(df['Preprocessed_Text'], df['Label_num'],
                                                   test_size=.2, random_state=123, stratify=df['Label_num'])
```

```python
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

## Machine Learning Model

### Naive Bayes

```python
# Create classifier
clf = Pipeline([
    ('vectorizer', TfidfVectorizer()),
    ('naive_bayes', (MultinomialNB()))
])
```

```python
# Model training
clf.fit(X_train, y_train)
```

```python
# Get prediction
y_pred = clf.predict(X_test)
```

```python
# Print score
print(accuracy_score(y_test, y_pred))
```

```python
# Print classification report
print(classification_report(y_test, y_pred))
```

## Random Fores

```python
# Create classifier
clr = Pipeline([
    ('vectorizer', TfidfVectorizer()),
    ('random_forest', (RandomForestClassifier()))
])
```

```python
# Model training
clr.fit(X_train, y_train)
```

```python
# Get prediction
y_pred = clr.predict(X_test)
```

```python
# Print score
print(accuracy_score(y_test, y_pred))
```

```python
# Print classification report
print(classification_report(y_test, y_pred))
```
